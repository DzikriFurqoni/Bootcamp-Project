## import library

```python
# Import libraries
import pickle
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from PIL import Image
```

Berikut adalah Libraries yang saya gunakan pada project ini

## load model

```python
model2 = load_model('model.h5')
```

## import gambar

```python
img = image.load_img('predict.jpg')
img
```

Mengambil sampel gambar random untuk digunakan sebagai prediksi, Dan menunjukan preview dari gambar tersebut.

```python
img = image.img_to_array(img) / 220
img = tf.image.resize(img, (220,220))
img = tf.expand_dims(img, axis=0)
```

Meresize ukuran gambar menjadi 220 * 220

## predict

```python
prediction = int(tf.round(model2.predict(x = img)).numpy()[0][0])

def classpredict(value):
    if value == 0:
        return "cat"
    elif value == 1:
        return "dog"

print('Image 1 Prediction : ', classpredict(prediction))
```

Setelah dilakukan Prediksi dapat kita lihat hasilnya menunjukan bahwa objek yang kita prediksi adalah cat.
